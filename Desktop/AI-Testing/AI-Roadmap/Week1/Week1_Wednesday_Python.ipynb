```python
# Lists — ordered, changeable, allow duplicates
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")  # add to end
fruits.remove("banana")  # remove by value
print(fruits[0])  # index from 0
print(fruits[-1])  # last item
print(fruits[1:3])  # slicing
print(fruits)
```

```python
# Starting data
watchlist = ["Inception", "The Matrix", "Interstellar", "Parasite", "Dune"]
watched = ["The Matrix", "Parasite"]

# Task 1 — Add "Oppenheimer"
watchlist.append("Oppenheimer")

# Task 2 — Remove movies already watched
watchlist = [movie for movie in watchlist if movie not in watched]

# Task 3 — Insert at position 2
watchlist.insert(2, "Everything Everywhere All at Once")

# Task 4 — Sort alphabetically
watchlist.sort()

# Task 5 — Print last 3 movies (no negative indexing)
last_three = watchlist[len(watchlist) - 3 :]
print(last_three)
```

```python
# Each movie is now [title, rating, year]
movies = [
    ["Inception", 8.8, 2010],
    ["Dune", 8.0, 2021],
    ["Oppenheimer", 8.9, 2023],
    ["Interstellar", 8.7, 2014],
]

# Sort this list by **rating**, highest first — without using a lambda (use `operator.itemgetter`)
from operator import itemgetter
movies.sort(key=itemgetter(1), reverse=True)

# Print only movies released **after 2015**
movies_after_2015 = [movie for movie in movies if movie[2] > 2015]

# Find the **average rating** using a list comprehension + `sum()` and `len()`
ratings = [movie[1] for movie in movies]
avg_rating = sum(ratings) / len(ratings)

print(avg_rating)
```

```python
# Dicts — key-value pairs, fast lookup
person = {"name": "Alice", "age": 30, "city": "Toronto"}
print(person["name"])  # access by key
person["email"] = "a@email.com"  # add new key
print(person.get("phone", "N/A"))  # safe get
for k, v in person.items():
    print(f"{k}: {v}")
```

```python
## Exercise: Movie database manager
# Starting data
movies = [
    {"title": "Inception",     "rating": 8.8, "year": 2010, "genre": "sci-fi"},
    {"title": "Dune",          "rating": 8.0, "year": 2021, "genre": "sci-fi"},
    {"title": "Oppenheimer",   "rating": 8.9, "year": 2023, "genre": "drama"},
    {"title": "Interstellar",  "rating": 8.7, "year": 2014, "genre": "sci-fi"},
]


# Task 1** — Add a `"watched"` key to every movie, set to `False` by default *(hint: loop through and update each dict)*

for movie in movies:
    movie["watched"] = False

print(movies)   

# Task 2** — Mark `"Inception"` and `"Dune"` as watched *(find them by title, update the value)*
for movie in movies:
    if movie["title"] in ["Inception", "Dune"]:
        movie["watched"] = True

print(movies)

# Task 3** — Build a new dict called `ratings_map` that maps each title → its rating
ratings_map = {movie["title"]: movie["rating"] for movie in movies}
print(ratings_map)

# Task 4** — Using `ratings_map`, safely look up the rating for `"Tenet"` — return `"Not in database"` if it doesn't exist *(one line, no `if` statement)*
tenet_rating = ratings_map.get("Tenet", "Not in database")
print(tenet_rating)

# Task 5** — Group movies by genre into a dict of lists:
genre_map = {}
for movie in movies:
    genre = movie["genre"]
    if genre not in genre_map:
        genre_map[genre] = []  # create list if genre not seen
    genre_map[genre].append(movie["title"])
print(genre_map)
```

```python
# Sets — unique values, fast membership test
a = {1, 2, 3, 2, 1}  # duplicates dropped
b = {3, 4, 5}
print(a)  # {1, 2, 3}
print(a & b)  # intersection
print(a | b)  # union
print(a - b)  # difference
print(2 in a)  # membership
```
